```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Create data with a non-linear pattern
X, y = make_regression(n_samples=100, n_features=1, bias=3, noise=15, random_state=42)

# Visualize the data
plt.scatter(X,y)
plt.show()

```

```python
## Splitting data for model traning and develpoment
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=42)

# Linear Regression Simple Model
model_1 = LinearRegression()
model_1.fit(X_train,y_train)

## Polynomial Regression more Complex Model
poly = PolynomialFeatures(degree=4)

X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
model_2 = LinearRegression()
model_2.fit(X_train_poly, y_train)
```

```python
# Prepare a wide range of X values to plot smooth lines
plot_x = np.linspace(X.min(), X.max(), 100)[:, np.newaxis]

# Predictions for the linear model
y_pred_1 = model_1.predict(plot_x)

# Predictions for the polynomial model (transformed plot_x)
y_pred_2 = model_2.predict(poly.transform(plot_x))

# Plot everything
plt.scatter(X, y, label="Data")
plt.plot(plot_x, y_pred_1, color='red', label="Linear Model")
plt.plot(plot_x, y_pred_2, color='yellow',  label="Polynomial Model")
plt.legend()
plt.show()
```
