* Year: The year of the flight (all records are from 2013)

* ArrDelay15: A binary indicator that arrival was delayed by more than 15 minutes (and therefore considered "late")
* Cancelled: A binary indicator that the flight was cancelled

## Checking For Outliers

* View the distribution and summary statistics for the `DepDelay` and `ArrDelay` columns.

```python
# Function to show the summary statistics and distribution for a column

def show_distribution(var_data):
    
    
    # Get Statistics
    min_val = var_data.min()
    max_val = var_data.max()
    mean_val = var_data.mean()
    med_val = var_data.median()
    mode_val = var_data.mode()[0]
    
    print(var_data.name,'\nMinimum: {:.2f}\nMaximun: {:.2f}\nMean: {:.2f}\nMedian: {:.2f}\nMode: {:.2f}'.format(min_val,max_val,mean_val,med_val,mode_val))
    
    
    # Create a figure for 2 subplots (2 rows, 1 columns)
    fig, ax = plt.subplots(2,1, figsize=(10,4))
    
    # Plot the histogram
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frequency')
    
    # Add lines for the mean, median, mode
    ax[0].axvline(x=min_val, color='black', linestyle='dashed', linewidth=2)
    ax[0].axvline(x=mean_val, color='cyan', linestyle='dashed', linewidth=2)
    ax[0].axvline(x=med_val, color='red', linestyle='dashed', linewidth=2)
    ax[0].axvline(x=mode_val, color='yellow', linestyle='dashed', linewidth=2)
    ax[0].axvline(x=max_val, color='black', linestyle='dashed', linewidth=2)
    
    # Plot Boxplot
    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel('Value')
    
    # Add a Title to the figure
    fig.suptitle(var_data.name)
    
    fig.show()
    
    
# Call the function for each delay field

delayFields = ['DepDelay','ArrDelay']

for col in delayFields:
    show_distribution(flight_df[col])
```

* There are outliers at the lower and upper ends of both the variables - particularly at the upper.

* Let's trim the data so that we include only rows where the values  for these fields are within the 1st and 90th percentile.

```python
# Trim Outliers for ArrDelay based on 1% and 90% percentiles

ArrDelay_01percentile = flight_df.ArrDelay.quantile(0.01)
ArrDelay_90percentile = flight_df.ArrDelay.quantile(0.90)

flight_df = flight_df[flight_df.ArrDelay < ArrDelay_90percentile]
flight_df = flight_df[flight_df.ArrDelay > ArrDelay_01percentile]

# Trim Outliers for DepDelay based on 1% and 90% percentiles

DepDelay_01percentile = flight_df.DepDelay.quantile(0.01)
DepDelay_90percentile = flight_df.DepDelay.quantile(0.90)

flight_df = flight_df[flight_df.DepDelay < DepDelay_90percentile]
flight_df = flight_df[flight_df.DepDelay > DepDelay_01percentile]

for col in delayFields:
    show_distribution(flight_df[col])
```

### **How do the carriers compare in terms of arrival delay performance?**

### **Which departure airport has the highest average departure delay?**

```python
departure_airport_group = flight_df.groupby(flight_df.OriginAirportName)

mean_departure_delays = pd.DataFrame(departure_airport_group['DepDelay'].mean()).sort_values('DepDelay', ascending=False)
mean_departure_delays.plot(kind = "bar", figsize=(16,10))
mean_departure_delays
```

### **Do late departures tend to result in longer arrival delays than on-time departures?**

```python
flight_df.boxplot(column='ArrDelay', by='DepDel15', figsize=(12,8))
```

### **Which route (from origin airport to destination airport) has the most late arrivals?**

```python
# Add route columns

routes = pd.Series(flight_df['OriginAirportName'] + ' > ' + flight_df['DestAirportName'])

flight_df = pd.concat([flight_df, routes.rename("Route")], axis=1)

# Group by Routes

route_group = flight_df.groupby(flight_df.Route)

pd.DataFrame(route_group['ArrDel15'].sum()).sort_values('ArrDel15', ascending=False)
```

### **Which route has the highest average arrival delay?**

```python
pd.DataFrame(route_group['ArrDelay'].mean()).sort_values('ArrDelay', ascending=False)
```
